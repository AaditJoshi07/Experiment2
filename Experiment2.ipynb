```python
listfruits=["apple","orange","cherry"]
print(listfruits)
```

```python
listfruits=["apple","orange","apple","cherry"] #duplicates are allowed in the list
print(listfruits)
```

```python
#finding the length of list

listfruits=["apple","orange","apple","cherry"] #duplicates are allowed in the list
print(listfruits)
print(len(listfruits))

list1=[2,2.5,True,"Aadit"]
print(list1)
print(len(list1))
```

```python
#appending the element

listfruits=["apple","orange","cherry"]
print(listfruits[0])
print(listfruits[-2])
listfruits.append("banana")
print(listfruits[0])
print(listfruits[-2])
print(listfruits)
```

```python
#inserting the element

listfruits=["apple","orange","cherry"]
listfruits.insert(1,"banana")
print(listfruits)
```

```python
#replacing the element

listfruits=["apple","orange","cherry"]
listfruits[1]="banana"
print(listfruits)
```

```python
#extending the element

list1=["apple","orange","cherry"]
list2=["mango","pineapple","banana"]

#method 1
x=list1 + list2
print(x)

#method 2
list1.extend(list2)
print(list1)
```

```python
list1=["apple","banana","cherry","kiwi","mango","pineapple"]
print(list1[2:5])
print(list1[2:])
print(list1[:4])
```

```python
#Create a list of 5 students names
#Display the first and last student
#Add a new student
#Remove any one student
#Sort the list alphabetically

listname=["raj","rohit","virat","josh","david"]
print(listname)
print(listname[0])
print(listname[4])
listname.append("smith")
print(listname)
listname.remove("virat")
print(listname)

listname.sort()
print(listname)
```

```python
#Create a list of student marks
#Display highest and lowest marks
#Calculate average marks
#Sort the marks in ascending order

listmarks=[45,67,34,89,80]
print(max(listmarks))
print(min(listmarks))
print(sum(listmarks))
print(len(listmarks))
print(sum(listmarks)/len(listmarks))
listmarks.sort()
print(listmarks)
```
